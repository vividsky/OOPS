```python
from matplotlib import pyplot as plt,style
style.use('ggplot')
x1 = [1,2,3]
y1 = [4,5,2]
x2 = [7,4,9]
y2 = [4,3,6]
plt.plot(x1,y1,'r',label = 'line1',linewidth = 5)
plt.plot(x2,y2, 'g', label = 'line2')
plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('First plot')
plt.show()
```
