```python
def StudentDetailsPie():
    '''
    Information in form of pie chart of Subjects
    myexplode: to make the specified subject(OOPS) info look distinctive
    list(Subjects.keys()): list carrying Subjects name, Names is a list carrying names of students
    plt.pie():marks: marks of student(Student_name) in different subjects
    plt.suptitle(): Title of the pie graph
    plt.show(): plot the graph
    '''
    i = Names.index(Student_name)
    marks = [Subjects[j][i] for j in Subjects.keys()]
    myexplode = [0, 0, 0, 0, 0]
    myexplode[list(Subjects.keys()).index(Subject)] = 0.1
    plt.pie(marks, labels = list(Subjects.keys()), explode = myexplode)
    plt.suptitle(Student_name, fontsize=15)
    return plt.show()
```

```python
def SubjectDetailsBar():
    plt.bar(Names, Subjects[Subject], color = 'c')
    plt.title(f"{Subject}'s Marks")
    plt.xlabel('----Names----')
    plt.ylabel('Marks')
    return plt.show()
```

```python
def StudentDetailsBar():
    i = Names.index(Student_name)
    marks = [Subjects[j][i] for j in Subjects.keys()]
    plt.bar(list(Subjects.keys()), marks, color = 'b')
    plt.title(f"{Student_name}'s Marks")
    plt.xlabel('----Subjects----')
    plt.ylabel('Marks')
    return plt.show()
```

```python
def marksComparision():
    marks_data = [] 
    for i in range(len(Names)):
        marks_data.append([Subjects[j][i] for j in Subjects.keys()])
    
    for j in range(len(Names)):
        plt.plot(list(Subjects.keys()), marks_data[j], marker = 'o',label = Names[j]) 

    plt.legend()
    plt.xlabel('----Subjects----')
    plt.ylabel('Marks')
    return plt.show()
```
